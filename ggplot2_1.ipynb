##Stacked area

```r
p.dat<-data.frame(step=row.names(grp.dat),grp.dat,stringsAsFactors=F)
p.dat<-melt(p.dat,id='step')
p.dat$step<-as.numeric(p.dat$step)

p=ggplot(p.dat,aes(x=step,y=value))

p1=p + geom_area(aes(fill=variable)) + theme(legend.position="bottom")+

  theme(legend.title=element_blank())+
  scale_fill_discrete(name="Experimental\nCondition")+
  xlab("Time") +
  ylab("Generation MW") +
  ggtitle("Yearly Generation") +
  theme(plot.title = element_text(size = rel(2)))
p1
```

##Scatter plot

```r
library(ggplot2)
excel=read.csv('/Users/vlasis/Desktop/wine.csv', header = TRUE)

data1=data.frame(excel)
str(data1)

ggplot(data1, aes(x=data1$Year, y=data1$Age)) +
  geom_point(shape=1) +   
  geom_smooth(method=lm)+
  xlab("x-axis") +
  ylab("y-axis") +
  ggtitle("Scatter") +
  theme(plot.title = element_text(size = rel(2)))




```

```r
ggplot(data1, aes(data1$Year)) + 
  geom_line(aes(y = data1$Age, colour = "red")) + 
  geom_line(aes(y = data1$AGST, colour = "green"))+
  theme(legend.position="bottom")+
  xlab("x-axis") +
  ylab("y-axis") +

  ggtitle("Scatter") +
  theme(plot.title = element_text(size = rel(2)))+
  # Remove title for all legends
  theme(legend.title=element_blank())
 
```
